## Serializers

     In Django Rest Framework the SERIALIZERS are responsible for converting complex data such as querysets and 
     model instances to native python datatypes (called serialization). Then that can be easily rendered into
     JSON, XML or other content types which is understandable by the Front-End
     
     SERIALIZERS are also responsible for DESERIALIZATION (which means converting the parsed data to complex types)
     
### Serializer Class

    - It is very similar to the Django Form and ModelForm class, and includes validation flags on the various fields
      such as required, max_length and default.
    - DRF provides a Serializer class which gives you a powerful generic way to control the output of your responses, 
      as well as a ModelSerializer class which provides a useful shourcut for creating serializers that deal with model
      instances and query sets.
